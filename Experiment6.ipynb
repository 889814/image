```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import google.colab.patches as patches
from google.colab.patches import cv2_imshow
```

```python
#Import color Image
image_path = '/content/download (1).png'

#Read color image
gray_image = cv2.imread(image_path)

cv2_imshow(gray_image)
```

```python
#Import color Image
image_path = '/content/face.jpg'

#Read color image
color_image = cv2.imread(image_path)
cv2_imshow(color_image)
```

```python
#detect the boundaries in the image
edges = cv2.Canny(color_image, 100, 200)

#Display the edges
plt.figure(figsize=(10, 5))
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.show
```

```python
#detect the boundaries in the image
edges = cv2.Canny(gray_image, 100, 200)

#Display the edges
plt.figure(figsize=(10, 5))
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.show
```

```python
gray_image = cv2.cvtColor(gray_image, cv2.COLOR_BGR2GRAY)
dct=cv2.dct(np.float32(gray_image))
plt.figure(figsize=(10, 5))
plt.imshow(dct, cmap='gray')
plt.title('DCT')
plt.show
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(color_image, scaleFactor=1.1, minNeighbors=5)

#Draw rectangles around the faces

for (x, y, w, h) in faces:
  cv2.rectangle(color_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.figure(figsize=(10, 5))
plt.imshow(color_image)
plt.title('Faces')
plt.show
```
