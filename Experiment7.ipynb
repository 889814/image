```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import google.colab.patches as patches
from google.colab.patches import cv2_imshow
import imghdr
import os
```

**For color image**

```python
#Import color Image
image_path = '/content/GettyImages-1907862843.tiff'

#Read color image
color_image = cv2.imread(image_path)
cv2_imshow(color_image)
```

```python
# Lossless compression
cv2.imwrite('/content/lossless.png', color_image)

png_path = '/content/lossless.png'
png_img = cv2.imread(png_path)
print("New Image type :",imghdr.what(png_path))
cv2_imshow(png_img)
```

```python
# Lossy Compression
cv2.imwrite('/content/lossy.jpeg', color_image, [cv2.IMWRITE_JPEG_QUALITY, -50])

jpeg_path = '/content/lossy.jpeg'
jpeg_img = cv2.imread(jpeg_path)
print("New Image type :",imghdr.what(jpeg_path))
cv2_imshow(jpeg_img)
```

```python
# Comparison
def plot_images(images, titles):
    """Plots a list of images with their corresponding titles."""
    fig, axes = plt.subplots(1, len(images), figsize=(15, 5))
    for i, (image, title) in enumerate(zip(images, titles)):
        axes[i].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # coz matplotlib expects the format to be RGB
        axes[i].set_title(title)
        axes[i].axis('off')
    plt.show()
plot_images([color_image,png_img,jpeg_img],['Original in tiff','Lossless compression in png','Lossy Compression in jpeg'])
```

```python
#Checking the size and type of image
print("Image Type for Original image:", imghdr.what(image_path))
print("Image Size for Original image:", os.path.getsize(image_path))
print("Image Shape for Original image:", color_image.shape)

#Checking the size and type of image
print("\nImage Type for Lossless image:", imghdr.what(png_path))
print("Image Size for Lossless image:", os.path.getsize(png_path))
print("Image Shape for Lossless image:", png_img.shape)

#Checking the size and type of image
print("\nImage Type for Lossy image:", imghdr.what(jpeg_path))
print("Image Size for Lossy image:", os.path.getsize(jpeg_path))
print("Image Shape for Lossy image:", jpeg_img.shape)
```
